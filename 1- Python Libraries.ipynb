## Matplotlib

```python
bins=10
alpha=0.5

setosa['sepal_length'].plot(kind='hist', label='setosa', bins=bins, alpha=alpha)
versicolor['sepal_length'].plot(kind='hist', label='versicolor', bins=bins, alpha=alpha)
virginica['sepal_length'].plot(kind='hist', label='virginita', bins=bins, alpha=alpha)

plt.title('Sepal length comparasion')

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1,2)

setosa.plot(x="sepal_length", y="sepal_width", kind="scatter",ax=ax[0],label='setosa',color='r')
versicolor.plot(x="sepal_length",y="sepal_width",kind="scatter",ax=ax[0],label='versicolor',color='b')
virginica.plot(x="sepal_length", y="sepal_width", kind="scatter", ax=ax[0], label='virginica', color='g')

setosa.plot(x="sepal_length", y="sepal_width", kind="scatter",ax=ax[1],label='setosa',color='r')
versicolor.plot(x="sepal_length",y="sepal_width",kind="scatter",ax=ax[1],label='versicolor',color='b')
virginica.plot(x="sepal_length", y="sepal_width", kind="scatter", ax=ax[1], label='virginica', color='g')

ax[0].set(title='Sepal comparasion ', ylabel='sepal-width')
ax[1].set(title='Petal Comparasion',  ylabel='petal-width')
ax[0].legend()
ax[1].legend()

plt.show()
```

```python
import plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
```

```python
# Iris-setosa
trace_setosa = go.Scatter3d(
                        x = setosa.sepal_length,
                        y = setosa.sepal_width,
                        z = setosa.petal_length,
                        mode = 'markers',
                        opacity = 0.7,
                        name = "setosa",
                        marker = dict(size = 3)
)

# Iris-versicolor
trace_versicolor = go.Scatter3d(
                        x = versicolor.sepal_length,
                        y = versicolor.sepal_width,
                        z = versicolor.petal_length,
                        mode = 'markers',
                        opacity = 0.7,
                        name = "versicolor",
                        marker = dict(size = 3)
)

# Iris-virginica
trace_virginica = go.Scatter3d(
                        x = virginica.sepal_length,
                        y = virginica.sepal_width,
                        z = virginica.petal_length,
                        mode = 'markers',
                        opacity = 0.7,
                        name = "virginica",
                        marker = dict(size = 3)
)

list_3d = [trace_setosa, trace_versicolor, trace_virginica]

fig_3d = go.Figure(data = list_3d)
iplot(fig_3d)
```

# NetworkX

```python
!pip install networkx
```

```python
import networkx as nx
```

```python
links = iris_df.corr().stack().reset_index()
links.columns = ['var1', 'var2','value']

# correlation
threshold = -1

# Keep only correlation over a threshold and remove self correlation (cor(A,A)=1)
links_filtered=links.loc[ (links['value'] >= threshold ) & (links['var1'] != links['var2']) ]
 
# Build your graph
G=nx.from_pandas_edgelist(links_filtered, 'var1', 'var2', edge_attr=True)

# Plot the network
nx.draw_circular(G, with_labels=True, node_color='orange', node_size=300, edge_color='red', linewidths=1, font_size=10)
```
